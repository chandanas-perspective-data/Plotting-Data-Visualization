### Histogram

    Histogram are the most common means of looking at a single variable.It is used to summarize discrete or continuous data that are measured on interval scale.
    

```python
df
```

```python
df['total_bill'].mean()
```

```python
fig=plt.figure(figsize=(5,5))
axes1=fig.add_subplot(1,1,1)
axes1.hist(df['total_bill'],bins=10,color='blue',orientation='vertical')
axes1.set_title("Histogram of  Total bills")
axes1.set_xlabel('frequency')
axes1.set_ylabel('total_bill')
```
